```python
# Grouping by year 
import warnings
yearly_bookings = df.groupby(['arrival_date_year'])['hotel'].count().reset_index(name='booking_count')

# Plotting the results
plt.figure(figsize=(10, 6))
sns.countplot(x='arrival_date_year', data=df)
plt.title('Yearly Bookings')
plt.xlabel('Year')
plt.ylabel('Booking Count')
plt.xticks(ticks=range(1, 4), labels=['2015', '2016', '2017'])
plt.show()
```

```python
# 3d. Marketing strategies based on booking lead time and special requests
# For customers with no special requests, target last-minute deals
# For customers with special requests, offer tailored promotions
import warnings
warnings.filterwarnings("ignore")
bins = [ 0,7, 30, 90, 365]
labels = ['Last-Minute', 'Short-Term', 'Mid-Term', 'Long-Term']
df['lead_time_category'] = pd.cut(df['lead_time'], bins=bins, labels=labels, right=False)

no_special_requests = df[df['total_of_special_requests'] == 0]
special_requests = df[df['total_of_special_requests'] > 0]

plt.figure(figsize=(5,5))
sns.countplot(x='lead_time_category', data=no_special_requests, palette='coolwarm')
plt.title('Last-Minute Deals for Customers with No Special Requests')
plt.legend(labels=['No Special Requests'])
plt.show()

plt.figure(figsize=(5,5))
sns.countplot(x='lead_time_category', data=special_requests, palette='muted')
plt.title('Targeted Promotions for Customers with Special Requests')
plt.legend(labels=['Special Requests'])
plt.show()
```

# TASK-5: Revenue Management

```python
# 5a. Average Daily Rate (ADR) variation over time

month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df['arrival_date_month'] = df['arrival_date_month'].map(month_map)

df = df.dropna(subset=['arrival_date_month'])

# Group by year and month to calculate the average ADR
adr_over_time = df.groupby(['arrival_date_year', 'arrival_date_month'])['adr'].mean().reset_index()

# Plotting ADR over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=adr_over_time, x='arrival_date_month', y='adr', hue='arrival_date_year', marker='o')
plt.title('Average Daily Rate (ADR) Over Time')
plt.xlabel('Month')
plt.ylabel('adr')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.show()
```

```python
# 5b.Identify pricing strategies that maximize revenue

adr_by_segment = df.groupby('market_segment')['adr'].mean().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=adr_by_segment.index, y=adr_by_segment.values, palette='viridis')
plt.title('Average Daily Rate (ADR) by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()

adr_by_channel = df.groupby('distribution_channel')['adr'].mean().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=adr_by_channel.index, y=adr_by_channel.values, palette='magma')
plt.title('Average Daily Rate (ADR) by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5c.Specific room types associated with higher ADR
adr_by_room_type = df.groupby('reserved_room_type')['adr'].mean().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=adr_by_room_type.index, y=adr_by_room_type.values, palette='Blues')
plt.title('Average Daily Rate (ADR) by Reserved Room Type')
plt.xlabel('Reserved Room Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()
```
